```python
#グランドトゥルースとオプティカルフローの動画ファイルによる可視化
import numpy as np
import cv2
import os
from google.colab.patches import cv2_imshow
from IPython.display import clear_output

# 推論結果の読み込み
predicted_flow = np.load('/content/drive/MyDrive/DL24/final/EventCamera/dl_lecture_competition/submission.npy.npy')

# グランドトゥルースのディレクトリパス
gt_flow_dir = '/content/drive/MyDrive/DL24/final/EventCamera/dl_lecture_competition/data/test/seq_1/flow_forward'

def load_gt_flow(gt_flow_file):
    flow_16bit = cv2.imread(gt_flow_file, cv2.IMREAD_ANYDEPTH|cv2.IMREAD_COLOR)
    flow_16bit = flow_16bit[:, :, ::-1] # BGR to RGB
    flow, valid2D = flow_16bit_to_float(flow_16bit)
    return flow, valid2D

def flow_16bit_to_float(flow_16bit):
    h, w, _ = flow_16bit.shape
    valid2D = flow_16bit[..., 2] == 1
    flow = np.zeros((h, w, 2), dtype=np.float32)
    flow[valid2D, 0] = (flow_16bit[valid2D, 0].squeeze() - 2**15) / 128
    flow[valid2D, 1] = (flow_16bit[valid2D, 1].squeeze() - 2**15) / 128
    return flow, valid2D

def flow_to_image(flow, max_flow=256):
    # オプティカルフローの大きさと角度を計算
    mag, ang = cv2.cartToPolar(flow[..., 0], flow[..., 1])

    # 角度を0-360度の範囲に正規化
    ang = ang * 180 / np.pi / 2
    ang = ang.astype(np.uint8)

    # 大きさを0-255の範囲に正規化
    mag = cv2.normalize(mag, None, 0, 255, cv2.NORM_MINMAX)
    mag = mag.astype(np.uint8)

    # HSV画像を作成
    hsv = np.zeros((flow.shape[0], flow.shape[1], 3), dtype=np.uint8)
    hsv[..., 0] = ang # 角度をHに対応
    hsv[..., 1] = 255 # 彩度は最大値
    hsv[..., 2] = mag # 大きさをVに対応

    # HSVからBGRに変換
    bgr = cv2.cvtColor(hsv, cv2.COLOR_HSV2BGR)

    return bgr

# グランドトゥルースの読み込みと可視化
gt_flow_files = sorted(os.listdir(gt_flow_dir))
for gt_flow_file in gt_flow_files:
    gt_flow_path = os.path.join(gt_flow_dir, gt_flow_file)
    gt_flow, gt_valid2D = load_gt_flow(gt_flow_path)

    print("pred_flow shape:", predicted_flow.shape)
    print("gt_flow shape:", gt_flow.shape)

    # pred_flowの形状を (height, width, 2) に変更
    if predicted_flow.ndim == 4:
        pred_flow = predicted_flow.transpose(0, 2, 3, 1)
        pred_flow = np.mean(pred_flow, axis=0)

    # gt_flowをpred_flowの形状に合わせてリサイズ
    if pred_flow.shape != gt_flow.shape:
        gt_flow = cv2.resize(gt_flow, (pred_flow.shape[1], pred_flow.shape[0]))

    # オプティカルフローを可視化用の画像に変換
    pred_img = flow_to_image(pred_flow)
    gt_img = flow_to_image(gt_flow)

    # 予測とGTを並べて表示
    combined_img = np.hstack((pred_img, gt_img))

    # 正解と判断した部分に緑色のバウンディングボックスを描画
    mask = np.linalg.norm(pred_flow - gt_flow, axis=2) < 5.0 # 閾値は適宜調整
    contours, _ = cv2.findContours(mask.astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        cv2.rectangle(combined_img, (x, y), (x+w, y+h), (0, 255, 0), 2) # 緑色のバウンディングボックス

    # 画像をリサイズ（オプショナル）
    combined_img = cv2.resize(combined_img, (1280, 480))

    # アニメーションで表示
    cv2_imshow(combined_img)
    clear_output(wait=True)
    cv2.waitKey(100)  # 100ミリ秒の遅延

cv2.destroyAllWindows()
```
